```python
import numpy as np
from implementations_leo import *
from proj1_helpers import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y, x, ids = load_csv_data("train.csv", sub_sample=True)
#y_t, x_t, ids_test = load_csv_data("test.csv", sub_sample=False)
```

```python
x[x == -999] = 0
```

```python
#Train and test with GD first for different gammas
#Results are OK, except diverges for gamma bigger than 10^{-5}
#I've tried to Raise Exception when loss in "nan', but haven't been able to 
initial_w = np.ones(30)/10
max_iters = 10000
ratio = 0.8
seed = 354

y_train, x_train, y_test, x_test = split_data(y,x, ratio, seed)

gammas = np.logspace(-7,-4,10)
losses = []

for gamma in gammas:
    print (gamma,"gamma")
    loss, w = least_squares_GD(y_train, x_train, initial_w, max_iters, gamma)
    print (loss, w, "rmse loss and weights")
    loss_test = cost_function(y_test, x_test,w)
    losses.append(loss_test)
    y_predicted = np.sign(np.dot(x_test,w))#Classification by sign of output
    print(sum(y_predicted == y_test)/len(y_test)*100. , "percentage of match")
    
print (losses,"losses")
plt.plot(gammas, losses)
```

```python
#Train and test with GD, but for polynomial basis
#As expected, takes much longer than linear basis - could potentially record execution time?
max_iters = 10000
ratio = 0.8
seed = 578
degree = 12
initial_w = np.ones(30*degree + 1)/10

y_train, x_train, y_test, x_test = split_data(y,x, ratio, seed)
xp_train = build_poly(x_train, degree,True )#No normalization, since its returning an error
xp_test = build_poly(x_test, degree, True)
print (xp_train.shape, xp_test.shape)

gammas = np.logspace(-7,-4,10)
losses = []

for gamma in gammas:
    print (gamma,"gamma")
    loss, w = least_squares_GD(y_train, xp_train, initial_w, max_iters, gamma)
    print (loss, w, "rmse loss and weights")
    loss_test = cost_function(y_test, xp_test,w)
    losses.append(loss_test)
    y_predicted = np.sign(np.dot(xp_test,w))#Classification by sign of output
    print(sum(y_predicted == y_test)/len(y_test)*100. , "percentage of match")
    
print (losses,"losses")
plt.plot(gammas, losses)
```
